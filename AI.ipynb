```python
#리스트 조작하기(append,insert,extend)
colors = ['red','green','gold']
print(colors)
#append(),리스트 마지막에 추가
colors.append('blue')
print("append=",colors)
#insert(), 리스트 원하는 곳 추가
colors.insert(1,'black')
print("insert=",colors)
#extend(),리스트 마지막에 추가
colors.extend(['white','gray'])
print("extend:",colors)
#리스트 마지막에 'yellow'추가
colors +=['yellow']
print("리스트에 +연산자 지원:", colors)
#리스트 마지막에 'r','e','d' 추가
colors += 'red'
print("리스트 연산에 문자열을 더하면", colors)
```

```python
#index(), 리스트에 어떤값이 어디에 있는지 확인하기
print(colors)
print("red의 위치:", colors.index('red'),"yellow의 위치 :", colors.index('yellow'))
```

```python
#count:해당값의 갯수 반환, pop:리스트의 마지막값이 제거
colors.append('red')
print(colors)
print("red의 개수:" , colors.count('red'))
print(colors.pop())
print(colors.pop())
print(colors)
```

```python
#remove: 특정위치 값 삭제, #sort,reverse 순,역방향 정렬
print(colors)
colors.remove('blue')
print(colors)
colors.sort()
print(colors)
colors.reverse()
print(colors)
```

```python
#knn 파이썬
#1. 데이터배열- 시각화
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

iris = load_iris()

print(iris)
print(iris.DESCR)

print(iris.data)
print(iris.feature_names)

print(iris.target)
print(iris.target_names)

df = pd.DataFrame(data = iris.data, columns = iris.feature_names)
df['target'] = iris.target

df['target'] = df['target'].map({0:"setosa", 1:"versicolor", 2:"virginica"})
print(df)

x_data = df.iloc[:,:-1]
y_data = df.iloc[:,[-1]]

#sns.pairplot(df, x_vars=["sepal length(cm)"], y_vars=["sepal width (cm)"], hue="target", height=5)


```

```python
#붓꽃 분류 코드
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(iris['data'], iris['target'], radom_state =0)

from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics

knn = KNeighborsClassifier(n_neighbors = 5)
knn.fit(X_train, y_train)

y_pred = knn.predict(X_test)
scores = metrics.accuracy_score(y_test, y_pred)
print(scores)

```

```python
#k-means 클러스터링

from sklearn import datasets
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans

iris = datasets.load_iris()

x = iris.data[:, :2]
y = iris.target

plt.scatter(x[:,0],x[:,1], c=y, cmap = 'gist_rainbow')
plt.xlabel('sepal Length', fontsize = 18)
plt.ylabel('sepal Width', fontsize = 18)

km = KMeans(n_clusters = 3, n_jobs = 4, random_state =21)
km.fit(x)

centers= km.cluster_centers_
print(centers)
```
